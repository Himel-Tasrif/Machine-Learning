## Log Loss Function

```python
np.log(0)
```

```python
epsilon = 1e-15
def log_loss(y_true, y_predicted):
    y_pred_new = [max(i,epsilon)for i in y_predicted]
    y_pred_new1 = [min(i,1-epsilon)for i in y_pred_new]
    y_pred_new2 = np.array(y_pred_new1)


    return -np.mean((y_true*np.log(y_pred_new2)) + ((1-y_true) * np.log(1-y_pred_new2)))
```

```python
# gradient descent

def gradient_descent(age, affordibility, y_true, epochs, loss_thresold):  # y_true is basically our truth value bought_insurance
    #w1,w2,bias
    w1 = w2 = 1
    bias = 0
    learning_rate = 0.5
    n = len(age)


    for i in range(epochs):
        y = w1*age + w2*affordibility + bias
        y_predicted = sigmoid_np(y)
        loss = log_loss(y_true, y_predicted)


        d_w1 = np.mean(age*(y_predicted - y_true)) # 1/n * (np.dot(np.transpose(age),(y_predicted - y_true)))

        d_w2 = np.mean(affordibility*(y_predicted - y_true)) # 1/n * (np.dot(np.transpose(affordibility),(y_predicted - y_true)))

        d_bias = np.mean((y_predicted - y_true))

        w1 = w1 - learning_rate*d_w1
        w2 = w2 - learning_rate*d_w2
        bias = bias - learning_rate*d_bias

        print(f'Epoch{i}-------w1: {w1},  w2: {w2}, bias: {bias}, loss: {loss}')

        if loss <= loss_thresold:
            break
    return w1,w2,bias
```
