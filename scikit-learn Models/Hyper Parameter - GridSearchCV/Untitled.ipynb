## Finding best model using GridSearchCV

```python
def find_best_model(x,y):

    algo = {
        'LinearRegression': {
            'model' : LinearRegression(),
            'param' : {
                'n_jobs' : [1,2,3,4]
            }
        },

        'Lasso': {
            'model' : Lasso(),
            'param' : {
                'alpha': [1, 2, 3],
                'max_iter': [1000, 2000],
                'random_state': [10, 20, 42]
            }
        },

        'RandomForestRegressor': {
            'model' : RandomForestRegressor(),
            'param' : {
                'n_estimators': [100,200,3000],
                'max_depth': [10,15,20],
                'criterion': ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
                'random_state': [1, 20, 42]
            }
        },

        'Support_Vec_R': {
            'model' : SVR(),
            'param' : {
                'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
                'gamma':['scale', 'auto'],
                'C': [1,2,3],
                'max_iter': [10,20,100]
            }
        },

        'GradientBoostingRegressor' : {
            'model': GradientBoostingRegressor(),
            'param': {
                'loss': ['squared_error', 'absolute_error', 'huber', 'quantile'],
                'n_estimators': [100, 300, 1000],
                'criterion': ['friedman_mse', 'squared_error'],
                'max_depth': [10,15,20],
                'random_state': [1,20,42]
            }
        },
    }


    score = []

    ss = ShuffleSplit(n_splits=5, test_size=0.1)

    for model_name, config in algo.items():
        gs = GridSearchCV(config['model'], config['param'], cv = ss, return_train_score = False)
        gs.fit(x, y)
        score.append({
            'Model_Name': model_name,
            'Best_Score': gs.best_score_,
            'Best_EST': gs.best_estimator_
        })


    return pd.DataFrame(score, columns = ['Model_Name', 'Best_Score', 'Best_EST'])


result = find_best_model(df2,df3)
result
```
