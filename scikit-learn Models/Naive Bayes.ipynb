# Part 1: predict titanic survival 

```python
import pandas as pd 
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis = 'columns',inplace= True)
```

```python
df.head()
```

```python
target = df.Survived
target
```

```python
train = df.drop('Survived',axis = 'columns')
train.head()
```

```python
dummy = pd.get_dummies(train.Sex)
dummy.head()
```

```python
train = pd.concat([train,dummy],axis = 'columns')
train.head()
```

```python
train.drop("Sex",axis = 'columns',inplace = True)
train.head()
```

```python
train.columns[train.isna().any()]
```

```python
train.Age = train.Age.fillna(train.Age.mean())
```

```python
train.columns[train.isna().any()]
```

```python
train.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train,target,test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_test[:5]
```

```python
model.predict(X_test[:10])
```

```python
y_test[:10]
```

```python
model.predict_proba(X_test[:10])
```

# Part 2: Spam Email detected

```python
import pandas as pd
hr = pd.read_csv("spam.csv")
hr
```

```python
dir(hr)
```

```python
hr.groupby('Category').describe()
```

```python
hr['new_cate'] = hr['Category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
hr
```

```python
hr1 = hr.drop('Category',axis = 'columns')
hr1.head()
# 1 means spam
# 0 means good
```

```python
x = hr1.Message
x
```

```python
y = hr1.new_cate
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bal = CountVectorizer()
x_train_count = bal.fit_transform(x_train.values)
x_train_count.toarray()[:6]
```

```python
bal.get_feature_names()
```

```python
bal
```

```python
from sklearn.naive_bayes import MultinomialNB
lol = MultinomialNB()
lol.fit(x_train_count,y_train)
```

```python
emails = [
    'Hey Rani , can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking , exclusive offer just for you . Do not miss this reward!'
]
emails_count = bal.transform(emails)
lol.predict(emails_count)
```
