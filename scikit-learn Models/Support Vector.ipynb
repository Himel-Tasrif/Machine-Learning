```python
df1 = pd.DataFrame(bf.data, columns = bf.feature_names)
```

```python
df1
```

```python
df1['target'] = bf.target
df1
```

```python
df1[df1.target == 1].head()
```

```python
df1['flower_name'] = df1.target.apply(lambda x : bf.target_names[x])
```

```python
df1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ir0 = df1[df1.target == 0]
ir1 = df1[df1.target == 1]
ir2 = df1[df1.target == 2]
```

```python
plt.title("Test 1")
plt.xlabel('sepal length (cm)', color = 'White')
plt.ylabel('sepal width (cm)', color = 'White')
plt.scatter(ir0['sepal length (cm)'],ir0['sepal width (cm)'],color = 'Green', marker = '+')
plt.scatter(ir1['sepal length (cm)'],ir1['sepal width (cm)'],color = 'Red', marker = '+')
```

```python
plt.title("Test 2")
plt.xlabel('sepal length (cm)', color = 'White')
plt.ylabel('sepal width (cm)', color = 'White')
plt.scatter(ir0['petal length (cm)'],ir0['petal width (cm)'],color = 'Green', marker = '+')
plt.scatter(ir1['petal length (cm)'],ir1['petal width (cm)'],color = 'Red', marker = '+')
```

```python
Y = df1.target
```

```python
Y
```

```python
X = df1.drop(['flower_name','target'],axis = 'columns')
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
len(x_test)
```

```python
model.predict(x_test)
```
