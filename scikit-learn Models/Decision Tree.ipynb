## Decision Tree

```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df
```

```python
target = df['salary_more_then_100k']
dr = df.drop('salary_more_then_100k',axis = 'columns')
```

```python
dr
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb_company = LabelEncoder()
lb_job = LabelEncoder()
lb_degree = LabelEncoder()
```

```python
dr['company_n'] = lb_company.fit_transform(dr['company'])
dr['job_n'] = lb_job.fit_transform(dr['job'])
dr['degree_n'] = lb_degree.fit_transform(dr['degree'])
```

```python
dr
```

```python
new_dr = dr.drop(['company','job','degree'],axis = 'columns')
new_dr
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(new_dr,target)
```

```python
model.predict([[2,2,1]])
```

```python
target
```

```python
model.score(new_dr,target)
```
