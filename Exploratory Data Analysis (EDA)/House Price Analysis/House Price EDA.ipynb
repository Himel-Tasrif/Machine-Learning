```python
test = pd.read_csv("test.csv")
test
```

```python
df = pd.concat([train,test])
df
```

```python
null_count = df.isna().sum()
null_columns = null_count[null_count > 0]
null_columns
```

```python
#               [expression for variable in iterable if condition]
df_with_null = [i for i in df.columns if df[i].isna().sum() > 1]

for i in df_with_null:
    print(i,np.round(df[i].isna().mean(), 4), '% missing values')

    
'''
A = 1,2,nan,4,5
A = f,f,t,f,f
    0,0,1,0,0
df[i].isna().mean() ------> sum of all true(1) and divide by total 
 (0 + 0 + 1 + 0 + 0) / 5 = 0.2 

this way it works
'''
```

```python
import seaborn as sns
```

## There are many missing values , we need to find the relation
## between missing values and SalePrice

```python
for i in df_with_null:
    df1 = df.copy()
    df1[i] = np.where(df1[i].isnull(),1,0)
    
    #all the null values converted 1 and non null values converted 0
    
    s = df1.groupby(i)['SalePrice'].median()
    #df1.groupby(i)['SalePrice'].median().plot().bar()
    plt.bar(s.index,s.values, color = ('green','blue'))
    plt.title(i)
    plt.show()
```

here we basically find the null(1) and non-null(0) values based on SalePrice(y-axis) and try to 
make decission which price is high null or non-null it play a major role

## Numarical Variables

```python
num_f = [i for i in df.columns if df[i].dtype != 'O'] # O mean objects

print('Numarical variables: ', len(num_f))

df[num_f].head()

'''
df[num_f] is the syntax for selecting a subset of columns from your DataFrame by passing a 
list of column names. 
This returns a new DataFrame that only contains the columns that are in the num_f list
'''
```

we have 4 datetime variables . we have extract information  from the datetime variable

```python
num_f
```

```python
dt = [i for i in num_f if 'Yr' in i or 'Year' in i]

dt
```

```python
for i in dt:
    print(i,df[i].unique())
```

```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
for i in dt:
    if i != 'YrSold':
        data = df.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[i] = data['YrSold'] - data[i]

        plt.scatter(data[i],data['SalePrice']) # find houses age 
        plt.xlabel(i) 
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## Numerical variables are usually of 2 type
## 1. Continous variable
## 2. Discrete Variables

discrete = [i for i in num_f if len(df[i].unique()) < 25 and i not in dt + ['Id']]
print("Discrete Variables Count: {}".format(len(discrete)))
```

```python
## Lets Find the realtionship between them and Sale PRice
# import matplotlib.cm as cm

for i in discrete:
    data = df.copy()
    data.groupby(i)['SalePrice'].median().plot.bar()
    # data.groupby(i)['SalePrice'].median().plot.bar(color = cm.get_cmap('rainbow')(np.arange(len(data[i].unique()))))
    
    plt.xlabel(i)
    plt.ylabel('SalePrice')
    plt.title(i)
    plt.show()
```

```python
continuous = [i for i in num_f if i not in discrete + dt + ['Id']]
print("Continuous feature Count {}".format(len(continuous)))
```

```python
df[continuous]
```

```python
for i in continuous:
    data = df.copy()
    data[i].hist(bins=25)
    plt.xlabel(i)
    plt.ylabel("Count")
    plt.title(i)
    plt.show()
```

```python
## We will be using logarithmic transformation


for i in continuous:
    data = df.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[i],data['SalePrice'])
        plt.xlabel(i)
        plt.ylabel('SalesPrice')
        plt.title(i)
        plt.show()
```

# Outlier

```python
for i in continuous:
    data = df.copy()
    if 0 in data[i].unique():
        pass
    else:
        data[i]=np.log(data[i])
        data.boxplot(column = i)
        plt.ylabel(i)
        plt.title(i)
        plt.show()
```
